### importing data

```python
import numpy as np 
import pandas as pd 
import requests
```

```python
url = "http://zims-en.kiwix.campusafrica.gos.orange.com/wikipedia_en_all_nopic/A/List_of_postal_codes_of_Canada:_M"
wiki_url = requests.get(url)
wiki_url
```

```python
wiki_data = pd.read_html(wiki_url.text)
wiki_data
```

```python
len(wiki_data), type(wiki_data)
```

```python
wiki_data = wiki_data[0]
wiki_data
```

```python
df = wiki_data[wiki_data["Borough"] != "Not assigned"]
df
```

```python
df = df.groupby(['Postcode']).head()
df
```

```python
df.Neighbourhood.str.count("Not assigned").sum()
```

```python
df = df.reset_index()
df
```

```python
df.drop(['index'], axis = 'columns', inplace = True)
df
```

```python
df.shape
```
